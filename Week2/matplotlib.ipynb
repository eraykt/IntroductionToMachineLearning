matplotlib grafik işlemleri sunar.

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.rand(1000) # rastgele 1000 veri oluşturduk
```

```python
plt.hist(data,bins=30)
plt.title("histogram")
plt.xlabel("Values")
plt.ylabel("Frequencies")
plt.show()
```

```python
sizes=[30,25,15,10,5,5]
plt.pie(sizes)
plt.title("Pie Chart")
plt.show()
```
